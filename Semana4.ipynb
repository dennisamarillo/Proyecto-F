```python
import sqlite3

conn = sqlite3.connect("proyecto.db")
cur = conn.cursor()

cur.executescript("""
DROP TABLE IF EXISTS pedido;
DROP TABLE IF EXISTS cliente;

CREATE TABLE cliente(
 id_cliente INTEGER PRIMARY KEY,
 nombre TEXT NOT NULL
);

CREATE TABLE pedido(
 id_pedido INTEGER PRIMARY KEY,
 id_cliente INTEGER,
 monto REAL,
 FOREIGN KEY(id_cliente)
 REFERENCES cliente(id_cliente)
);
""")
cur.execute("INSERT INTO cliente VALUES (20, 'Diego Luna')")
cur.execute("INSERT INTO cliente VALUES (30, 'Ricardo Mendoza')")
cur.execute("INSERT INTO cliente VALUES (40, 'Kathy Saenz')")
cur.execute("INSERT INTO cliente VALUES (50, 'Juan Perez')")
cur.execute("INSERT INTO pedido VALUES (1001, 20, 150.75)")
cur.execute("INSERT INTO pedido VALUES (1002, 30, 230.00)")
cur.execute("INSERT INTO pedido VALUES (1003, 20, 80.50)")
cur.execute("INSERT INTO pedido VALUES (1004, 40, 500.00)")
cur.execute("INSERT INTO pedido VALUES (1005, 30, 95.20)")
cur.execute("INSERT INTO pedido VALUES (1006, 50, 105.20)")


conn.commit()

print("\n--- Tabla: cliente ---")
print("ID | Nombre")
cur.execute("SELECT * FROM cliente")
for id_cli, nombre in cur.fetchall():
    print(f"{id_cli:<2} | {nombre}")

print("\n--- Tabla: pedido ---")
print("ID Pedido | ID Cliente | Monto")
cur.execute("SELECT * FROM pedido")
for id_ped, id_cli, monto in cur.fetchall():
    print(f"{id_ped:<9} | {id_cli:<10} | {monto}")

conn.close()
```

a) ¿Qué representan tus tablas principales dentro del negocio elegido?

Respuesta:

Las tablas representan dos entidades clave del negocio: la tabla "cliente" captura los datos de cada cliente, como su identificador y nombre; mientras que la tabla "pedido" registra las transacciones, vinculando cada compra al cliente y al monto de la operación.

b) ¿Qué relación implementaste con la FOREIGN KEY (1:1, 1:N o N:M) y por qué?

Respuesta:

La relación establecida es de uno a muchos (1:N). Esto permite que un cliente agrupe múltiples pedidos, pero cada pedido se asocia a un único cliente. Esta estructura es funcional para hacer seguimiento de las compras por cliente, evitando duplicidad de información.

c) Menciona una decisión de diseño que tomaste para evitar datos duplicados o desordenados.

Respuesta:

Una decisión de diseño fue normalizar los datos separando las tablas. La clave foránea (id_cliente) garantiza que la información del cliente no se repita en cada pedido, asegurando coherencia, evitando redundancias y facilitando el mantenimiento y la integridad de los datos.
